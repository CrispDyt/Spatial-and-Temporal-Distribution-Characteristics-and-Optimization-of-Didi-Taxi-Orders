```python
time_total=[]
for i in range(2712):
    time_total.append(driver_indent_time.xs(index_fresh[i],level='indent').head(1).index.values[0])
```

```python
from datetime import datetime
time_fin=[]
for i in range(2712):
    ts = (time_total[i] - np.datetime64('1970-01-01T00:00:00Z')) / np.timedelta64(1, 's')
    a=datetime.utcfromtimestamp(ts).strftime("%Y-%m-%d %H:%M:%S")
    time_fin.append(int(a[11])*10+int(a[12]))
```

```python
#22:00 - 4 :00
work_rush_time_total=0
##4:00 - 10 :00
day_time_total=0
## 10：00 - 16：00
backhome_time_total=0
## 16：00 - 22 ：00
night_total=0
for i in range(2712):
    if 4<= time_fin[i] <10:
        work_rush_time_total +=1
    elif 10<= time_fin[i] <16:
        day_time_total +=1
    elif 16<= time_fin[i] <22:
        backhome_time_total +=1
    else:
        night_total +=1
print(work_rush_time_total,day_time_total,backhome_time_total,night_total)
```
