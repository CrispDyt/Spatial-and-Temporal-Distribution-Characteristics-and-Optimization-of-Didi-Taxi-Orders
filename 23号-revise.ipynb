```python
from datetime import datetime
time_fin=[]
for i in range(2626):
    ts = (time_total[i] - np.datetime64('1970-01-01T00:00:00Z')) / np.timedelta64(1, 's')
    a=datetime.utcfromtimestamp(ts).strftime("%Y-%m-%d %H:%M:%S")
    time_fin.append(int(a[11])*10+int(a[12]))
```
